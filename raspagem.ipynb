# Gera planilhas com o links dos produtos


```python
import requests
from bs4 import BeautifulSoup
import openpyxl

busca = input("Produto: ")
url = "https://www.magazinevoce.com.br/magazinematapreco/busca/"+busca+"?&sortOrientation=desc&sortType=soldQuantity"
nome_arquivo = busca
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    product_cards = soup.find_all('li', class_='sc-heIBml')
    
    dados_magazine = []

    for card in product_cards:
        
        # Nome
        try:
            title = card.find('h2', {'data-testid': 'product-title'}).text.strip()
        except:
            title = "N/A"
            
        # Imagem
        try:
            image = card.find('img', {'data-testid': 'image'})['src']
        except:
            image = "N/A"
            
        # Preço original
        try:
            original_price = card.find('p', {'data-testid': 'price-original'}).text.strip()
        except:
            original_price = "N/A"
            
        # Preço com Desconto
        try:
            discount_price = card.find('p', {'data-testid': 'price-value'}).text.strip()
        except:
            discont_price = "N/A"
            
        #Link
        try:
            link = "https://www.magazinevoce.com.br" + card.find('a')['href']
        except:
            link = "N/A"

        dados_magazine.append([title, link, image, original_price, discount_price])

    # Criar um novo arquivo Excel
    workbook = openpyxl.Workbook()

    # Selecionar a planilha ativa
    sheet = workbook.active

    # Adicionar cabeçalhos às colunas
    sheet['A1'] = 'Nome'
    sheet['B1'] = 'Link'
    sheet['C1'] = 'Imagem'
    sheet['D1'] = 'Preço Original'
    sheet['E1'] = 'Preço com Desconto'

    # Adicionar os dados à planilha
    for row_idx, row_data in enumerate(dados_magazine, start=2):
        for col_idx, value in enumerate(row_data, start=1):
            sheet.cell(row=row_idx, column=col_idx, value=value)

    # Salvar o arquivo Excel
    workbook.save(f'{nome_arquivo}.xlsx')

    print(f"Dados salvos com sucesso no arquivo '{nome_arquivo}.xlsx'.")
else:
    print(f"Erro ao acessar a página. Código de status: {response.status_code}")
```
